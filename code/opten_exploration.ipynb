**location, industry aggregation**

Create a table for location, industry and number of firms, total number of employees.<br>
Location could be region, district, street or smaller.<br>
Industry categorization is available as 5 digit NACE codes in the main_activity_code column.<br>

```python
# opten -- firm registry
rdf = pd.read_csv("../data/opten_ENG/opten_firm_registry.csv", sep=";", encoding="utf-8-sig")
```

```python
# example -- lets choose district 9 in Budapest
test_df = rdf[(rdf["hq_city"]=="Budapest")]
test_df["district"] = test_df["hq_postcode"].astype(str).str[1:3]
test_df = test_df[test_df["district"] == "09"]

# lets focus on firms with min 2 employees
test_df = test_df[test_df["employees_2019"] >= 2]

# keep 2 digit industry codes (NACE codes)
test_df["NACE2"] = (test_df["main_activity_code"] // 1000).astype(int)
test_df["NACE3"] = (test_df["main_activity_code"] // 100).astype(int)
test_df["NACE4"] = (test_df["main_activity_code"] // 10).astype(int)
```

```python
# aggregation example
location = "hq_street_name"
industry_category = "NACE3"
agg_df = (
    test_df
    .groupby([location, industry_category])
    .agg(
        nr_firms = pd.NamedAgg("firm_id_opten", "nunique"),
        total_emp = pd.NamedAgg("employees_2019", "sum")
    )
    .reset_index()
)

agg_df
```
